```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
files = glob.glob("*.csv")
```

```python
pd.read_csv(files[0])
```

```python
plt.figure(figsize=(10,5))
for f in files:
    name = f.split(".")[0]
    df = pd.read_csv(f)
    df.sort_values(by = "VOLTAGE", inplace = True)
    plt.plot(df["VOLTAGE"], df["CURRENT"], label = name)
plt.legend(loc = [1.2,0.5])
    
```
