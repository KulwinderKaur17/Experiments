```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Heating

```python
df = pd.read_csv("data2.csv")
df
```

```python
df["Dielectric"] = df["Capacitance"]/0.349
```

```python
plt.figure(figsize = (10, 8))
plt.plot(df["Temperature"], df["Dielectric"])
plt.scatter(df["Temperature"], df["Dielectric"])
plt.xlabel("Temperature in $^O C$", fontsize=15)
plt.ylabel("Dielectric Constant $(\epsilon)$", fontsize=15)
plt.title("Temperature Vs Dielectric Constant", fontsize=15)
plt.grid()
plt.savefig("DC2.jpg")
```

```python
df.columns = ["Temperature in Degree Celcius", "Capacitance in pF", "Dielectric Constant"]
df
```

```python
df.to_html("Data2.html")
```
